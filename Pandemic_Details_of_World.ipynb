<h2>Basic Exploration</h2>

<h6>Rows and Columns</h6>

```python
# Read Data and check the number of rows and columns
df = pd.read_csv("pandemic_details_of_world.csv")

# head() allows you to view the first few rows of the data.
df.head().style.set_properties(**{"background-color": "#333333","color":"#F5F5F5","border": "1.5px solid black"})
```

```python
# The shape attribute of a pandas DataFrame is used to retrieve the dimensions of the DataFrame. 
# It returns a tuple containing the number of rows and columns in the DataFrame.
num_rows, num_cols = df.shape
print(f'Number of rows: {num_rows}\nNumber of columns: {num_cols}')
```

<h6>Data Type</h6>

```python
# Get the Data Type of each column.
column_name = list(df.columns.values)
column_dtype = pd.Series(df[column_name].dtypes)
column_dtype
```

<h6>Check for duplicate rows</h6>

```python
contains_duplicate_rows = df.duplicated().any()
contains_duplicate_rows
```

<h2>Discover</h2>

<h6>WordCloud for Disease</h6>

```python
disease_data = df2['Disease']
disease_data
```

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='black').generate('\n'.join(disease_data.dropna().values))

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
